```python
import numpy as np
import pandas as pd
```

```python
users_list = pd.read_table('data/ml-1m/users.dat',
                      sep='::', header=None, 
                      names=['user_id', 'gender', 'age', 'occupation', 'zip'], engine='python')

ratings_list = pd.read_table('data/ml-1m/ratings.dat',
                        sep='::', header=None, 
                        names=['user_id', 'movie_id', 'rating', 'timestamp'], engine='python')

movies_list = pd.read_table('data/ml-1m/movies.dat',
                       sep='::', header=None, 
                       names=['movie_id', 'title', 'genres'], engine='python')
```

```python
movies_list.head()
```

```python
ratings_list.head()
```

```python
movies_list = pd.concat([movies_list, movies_list.genres.str.get_dummies(sep='|')], axis=1)  
movies_list.head(5) 
```

```python
for index, row in movies_list.iterrows():
    main_g = row['genres'].split("|")[0]
    movies_list.loc[index , main_g] = 2
```

```python
movies_list.head(5)
```

```python
geners = list(movies_list)[3:]
geners
```

```python
movies_ratings = pd.merge(ratings_list,movies_list,on='movie_id', how='left')
movies_ratings.head()
```

```python
movieslens = pd.merge(movies_ratings,users_list,on='user_id',how='left')
```

```python
movieslens.head()
```

```python
movieslens_sl2 = movieslens[:50000]
movieslens_sl2
```

```python
movieslens_sl22 = movieslens_sl2.apply(lambda row: -1*(row[geners]/row['rating']) if row['rating'] < 3 
     else row[geners]*row['rating'], axis=1)
```

```python
movieslens_sl333 = pd.concat([movieslens_sl2[movieslens_sl2.columns[0:1]], movieslens_sl22], axis=1)
```

```python
user_preference = movieslens_sl333.groupby(['user_id']).mean()
```

```python
#in production you would use np.dot instead of writing your own dot product function.
def dot_product(vector_1, vector_2):  
    return sum([ i*j for i,j in zip(vector_1, vector_2)])

def get_movie_score(movie_features, user_preferences):  
    return dot_product(movie_features, user_preferences)
```

```python
toy_story_user_predicted_score = dot_product(toy_story_features, user_4_preference)  
toy_story_user_predicted_score  
```

```python
def get_movie_recommendations(user_preferences, n_recommendations):  
    #we add a column to the movies_df dataset with the calculated score for each movie for the given user
    movies_list['score'] = movies_list[geners].apply(get_movie_score, 
                                                           args=([user_preferences]), axis=1)
    return movies_list.sort_values(by=['score'], ascending=False)[['title','genres','score']][:n_recommendations]
```
